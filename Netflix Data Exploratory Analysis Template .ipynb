### <font color=green> Exploratory Analysis
### <font color=green> Distribution of Shows

```python
# Distribution of Category (Movies vs Shows)
category_distribution = data['Category'].value_counts()

import matplotlib.pyplot as plt 

category_distribution.plot(kind='bar', color=['red', 'blue'])
plt.title('Distribution of Content Types on Netflix')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

### <font color=green> Distribution of content by country

```python
country_distribution = data['Country'].dropna().str.split(', ').explode().value_counts().head(10)

country_distribution.plot(kind='bar', color='green')
plt.title('Top 10 Countries Contributing Content on Netflix')
plt.xlabel('Country')
plt.ylabel("Number of Shows")
plt.xticks(rotation=45)
plt.show()
```
